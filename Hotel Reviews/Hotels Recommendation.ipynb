### LOCATION BASED RECOMMENDATION SYSTEM

The column Address, city, province, longitude, latitude, and Country reveals the location of the hotel.

```python
# convert the text to lowercase 
hotel_loc['address'] = hotel_loc['address'].str.lower() 
hotel_loc['address'] = hotel_loc['address'].str.replace(" ", "")
```

```python

# convert the text to lowercase
hotel_loc['city'] = hotel_loc['city'].str.lower() 
hotel_loc['province'] = hotel_loc['province'].str.lower()
```

```python
hotel_loc.duplicated().sum()
```

```python
hotel_loc.head()
```

##### DROP DUPLICATES

```python
hotel_loc = hotel_loc.drop_duplicates()
```

```python
hotel_loc = hotel_loc.reset_index()
```

```python
hotel_loc.drop('index', axis = 1, inplace = True)
```

```python
hotel_loc.head()
```

##### JOIN THE LOCATION VARIABLES TOGETHER

```python

# Join the location variables into a sentence
hotel_loc['location'] = hotel_loc['address'] + " " + hotel_loc['city'] + " " + hotel_loc['province'] + " " + str(hotel_loc['latitude']) + " " + str(hotel_loc['longitude'])
```

```python
hotel_loc.head()
```

##### VECTORIZE THE DATASET

```python
# import necessary libraries 
from sklearn.feature_extraction.text import TfidfVectorizer 

# instantiate 
tfidf = TfidfVectorizer(analyzer = 'word', ngram_range = (1, 2), min_df = 0) 

# fit and transform
tfidf_vec = tfidf.fit_transform(hotel_loc['location'])
```

```python
tfidf_vec.shape
```

##### COMPUTE SIMILARITY SCORES

```python
# import necessary library 
from sklearn.metrics.pairwise import cosine_similarity 

# compute
cos_sim = cosine_similarity(tfidf_vec)

cos_sim[5:10, 5:10]
```

##### GET RECOMMENDATION

```python
hotel_names = hotel_loc['name']
indices = pd.Series(hotel_loc.index, index = hotel_loc['name'])
```

```python

# get recommendation function 
def get_recommendation(hotel_name):
    idx = indices[hotel_name]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:10]
    hotel_loc_indices = [i[0] for i in sim_scores]
    return pd.DataFrame(hotel_names.iloc[hotel_loc_indices])
```
