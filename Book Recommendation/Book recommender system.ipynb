```python

# lower the case of the author name
books['Book-Author'] = books['Book-Author'].str.lower() 

# join the author's name by removing the space 
books['Book-Author'] = books['Book-Author'].str.replace(' ', '')
```

```python
books['Publisher'] = books['Publisher'].str.lower()
```

##### DROP DUPLICATES

```python

# duplicates 
books.duplicated(subset = 'Book-Title').sum()
```

```python
# drop duplicates 
books = books.drop_duplicates(subset = 'Book-Title')
```

```python
books.duplicated(subset = 'Book-Title').sum()
```

### SAMPLE 50000 BOOKS RANDOMLY

```python
rand_books = books.sample(25000, replace = False, random_state = 2104)
```

```python
rand_books.head()
```

```python

# reset index
rand_books = rand_books.reset_index()
```

```python
# drop index column
rand_books = rand_books.drop('index', axis = 1)
```

```python
rand_books.head(3)
```

```python
# concatenate the features
rand_books['content'] = rand_books['Book-Author'] + " " + rand_books['Publisher']
```

```python
rand_books.head()
```

### DROP FEATURES THAT ARE NOT NEEDED

```python
rand_books.drop(columns = ['Image-URL-L', 'Image-URL-M', 'Image-URL-S', 'Year-Of-Publication', 'ISBN'], axis = 1, inplace = True)
```

```python
rand_books.head()
```

### TEXT VECTORIZATION

```python
# import necessary libraries 
from sklearn.feature_extraction.text import TfidfVectorizer 

tfidf = TfidfVectorizer(analyzer = 'word', ngram_range = (1, 2), stop_words = 'english') 

books_mat = tfidf.fit_transform(rand_books['content'])
```

```python
books_mat.shape
```

### COMPUTE COSINE SIMILARITY

```python
# import necessary libraries 
from sklearn.metrics.pairwise import cosine_similarity 

# instantiate 
cos_sim = cosine_similarity(books_mat)
```

```python
cos_sim[4:8, 4:8]
```

##### GET RECOMMENDATIONS

```python
titles = rand_books['Book-Title']
indices = pd.Series(rand_books.index, index = rand_books['Book-Title'])
```

```python
def get_recommendation(title):
    idx = indices[title]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return pd.DataFrame(titles.iloc[movie_indices])
```

```python
get_recommendation('Curtain Call')
```
